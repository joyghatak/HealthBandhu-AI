# Phase 3 : Emergency Detection Layer

This layer identifies potentially life-threatening symptom patterns
before disease prediction.

Emergency cases receive immediate priority and warning scores.

Priority:

Emergency Screening
↓
Disease Prediction
↓
Clinical Report Generation

```python
EMERGENCY_SYMPTOMS = {
    "shortness of breath": 5,
    "sharp chest pain": 5,
    "irregular heartbeat": 5,
    "breathing fast": 4,
    "chest tightness": 4,
    "throat swelling": 4,
    "fainting": 5,
    "seizures": 5
}

print("=" * 60)
print("EMERGENCY DETECTION SYSTEM INITIALIZED")
print("=" * 60)

for symptom, score in EMERGENCY_SYMPTOMS.items():
    print(f"{symptom} -> Risk Score {score}")
```

```python
# =====================================================
# EMERGENCY SCORING FUNCTION
# =====================================================

def calculate_emergency_score(symptom_vector):

    total_score = 0

    for symptom, weight in EMERGENCY_SYMPTOMS.items():

        if symptom in symptom_vector.index:

            if symptom_vector[symptom] == 1:

                total_score += weight

    return total_score
```

```python
# =====================================================
# EMERGENCY LEVEL CLASSIFIER
# =====================================================

def classify_emergency(score):

    if score >= 10:
        return "CRITICAL"

    elif score >= 5:
        return "HIGH RISK"

    elif score >= 2:
        return "MODERATE RISK"

    else:
        return "LOW RISK"
```

```python
# =====================================================
# EMERGENCY ANALYSIS ON TRAINING DATA
# =====================================================

emergency_scores = X_train.apply(
    calculate_emergency_score,
    axis=1
)

print("Maximum Emergency Score:",
      emergency_scores.max())

print("Average Emergency Score:",
      round(emergency_scores.mean(),2))

print("Critical Cases:",
      (emergency_scores >= 10).sum())
```

# Phase 4 : Rule-Based Clinical Layer

This layer automatically discovers important symptom patterns
for diseases and provides interpretable rule-based reasoning
before machine learning prediction.

```python
# =====================================================
# DISEASE SYMPTOM PROFILES
# =====================================================

TOP_DISEASES = (
    df_filtered["diseases"]
    .value_counts()
    .head(20)
    .index
)

disease_profiles = {}

for disease in TOP_DISEASES:

    subset = df_filtered[
        df_filtered["diseases"] == disease
    ]

    symptom_strength = (
        subset.drop(
            columns=["diseases"]
        )
        .mean()
        .sort_values(
            ascending=False
        )
    )

    disease_profiles[disease] = symptom_strength.head(10)

print(
    "Profiles Generated:",
    len(disease_profiles)
)
```

```python
# =====================================================
# SAMPLE CLINICAL RULES
# =====================================================

for disease, profile in list(
    disease_profiles.items()
)[:5]:

    print("\n")
    print("="*60)

    print("Disease:", disease)

    print("="*60)

    print(
        profile.index.tolist()
    )
```

```python
# =====================================================
# RULE MATCHING ENGINE
# =====================================================

def rule_match(patient_vector):

    scores = {}

    active_symptoms = set(
        patient_vector[
            patient_vector == 1
        ].index
    )

    for disease, profile in disease_profiles.items():

        disease_symptoms = set(
            profile.index
        )

        overlap = len(
            active_symptoms &
            disease_symptoms
        )

        scores[disease] = overlap

    return sorted(
        scores.items(),
        key=lambda x: x[1],
        reverse=True
    )[:5]
```

```python
# =====================================================
# RULE ENGINE TEST
# =====================================================

patient = X.iloc[0]

rule_predictions = rule_match(
    patient
)

print(
    "Top Rule-Based Predictions"
)

for disease, score in rule_predictions:

    print(
        disease,
        "->",
        score
    )
```

# Phase 6 : Deep Learning Layer

This phase trains a neural network for
large-scale multiclass disease classification.

Input Features : 377 Symptoms
Output Classes : 754 Diseases

```python
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
```

```python
print(X_train.shape)
print(len(label_encoder.classes_))
```

```python
# =====================================================
# DNN DATA PREPARATION
# =====================================================

import numpy as np

X_train_nn = X_train.astype("float32")
X_test_nn = X_test.astype("float32")

y_train_nn = y_train
y_test_nn = y_test

print("Training Shape :", X_train_nn.shape)
print("Testing Shape  :", X_test_nn.shape)
```

```python
# =====================================================
# DNN MODEL
# =====================================================

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization

NUM_CLASSES = len(label_encoder.classes_)

dnn_model = Sequential([

    Dense(
        512,
        activation="relu",
        input_shape=(377,)
    ),

    BatchNormalization(),

    Dropout(0.30),

    Dense(
        256,
        activation="relu"
    ),

    BatchNormalization(),

    Dropout(0.30),

    Dense(
        128,
        activation="relu"
    ),

    Dense(
        NUM_CLASSES,
        activation="softmax"
    )
])

dnn_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

dnn_model.summary()
```

```python
# =====================================================
# DNN TRAINING
# =====================================================

from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor="val_accuracy",
    patience=3,
    restore_best_weights=True
)

history = dnn_model.fit(
    X_train_nn,
    y_train_nn,
    validation_split=0.10,
    epochs=20,
    batch_size=1024,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# =====================================================
# DNN EVALUATION
# =====================================================

loss, accuracy = dnn_model.evaluate(
    X_test_nn,
    y_test_nn,
    verbose=0
)

print(
    f"DNN Accuracy: {accuracy:.4f}"
)
```

# Phase 10 : Confidence Layer

This phase evaluates prediction certainty
and categorizes diagnostic confidence.

```python
# =====================================================
# CONFIDENCE CLASSIFIER
# =====================================================

def confidence_level(prob):

    if prob >= 0.90:
        return "VERY HIGH"

    elif prob >= 0.75:
        return "HIGH"

    elif prob >= 0.50:
        return "MODERATE"

    else:
        return "LOW"
```

```python
# =====================================================
# CONFIDENCE ANALYSIS
# =====================================================

level = confidence_level(
    confidence
)

print(
    "Confidence Level:",
    level
)
```

# Phase 11 : Clinical Assistant

This layer converts AI predictions into
human-readable clinical recommendations.

```python
# =====================================================
# CLINICAL ASSISTANT
# =====================================================

def clinical_assistant(
    disease,
    confidence,
    emergency_level
):

    print("="*60)
    print("CLINICAL ASSISTANT")
    print("="*60)

    print(f"Disease Prediction : {disease}")

    print(
        f"Confidence : {confidence*100:.2f}%"
    )

    print(
        f"Emergency Level : {emergency_level}"
    )

    print("\nRecommendation:")

    if emergency_level == "CRITICAL":

        print(
            "Immediate medical evaluation recommended."
        )

    elif confidence >= 0.90:

        print(
            "Prediction highly reliable."
        )

    elif confidence >= 0.75:

        print(
            "Prediction reliable but requires confirmation."
        )

    else:

        print(
            "Further clinical assessment recommended."
        )
```
